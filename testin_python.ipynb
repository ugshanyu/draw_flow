```python
def add(input_1, input_2):
    return input_1 + input_2

def run_graph(data):
    nodes = data['drawflow']['Home']['data']
    outputs = {}

    # First, simulate input nodes
    for node_id, node_info in nodes.items():
        if node_info['name'] == 'input':
            # Simulating an input node by assigning a static value
            outputs[node_id] = 5  # Modify to dynamic if necessary

    # Next, execute other nodes
    for node_id, node_info in nodes.items():
        if node_info['name'] == 'add':
            inputs = []
            for input_key, input_value in node_info['inputs'].items():
                # Assuming there's only one connection per input
                source_node = input_value['connections'][0]['node']
                inputs.append(outputs[source_node])
            outputs[node_id] = add(*inputs)

    # Output the results from all nodes
    for node_id, output in outputs.items():
        print(f"Node {node_id} output: {output}")

```

```python
def handle_variable(node_data):
    # Handle the 'variable' type node
    if 'data' in node_data:
        method = node_data['data'].get('method', '')
        valueholder = node_data['data'].get('valueHolder', '')
        key = node_data['data'].get('key', '')
        return {
            "method": method,
            "valueholder": valueholder,
            "key": key
        }

def handle_llmGenerate(node_data):
    # Handle the 'llmGenerate' type node
    if 'data' in node_data:
        system_prompt = node_data['data'].get('systemPrompt', '')
        user_prompt = node_data['data'].get('userPrompt', '')
        return {
            "systemPrompt": system_prompt,
            "userPrompt": user_prompt
        }

def process_nodes(json_data):
    nodes = json_data['drawflow']['Home']['data']
    results = []

    # Define a dispatcher dictionary to call the right function based on the class type
    dispatcher = {
        'variable': handle_variable,
        'llmgenerate': handle_llmGenerate  # Use lowercase keys for case-insensitive matching
    }

    for node_key, node in nodes.items():
        node_class = node['class'].lower()  # Convert class to lowercase to match dispatcher keys
        if node_class in dispatcher:
            # Call the appropriate function based on the node class
            result = dispatcher[node_class](node['data'])  # Pass the correct part of the node to the handler
            if result:
                results.append(result)

    return results


# Example usage:
extracted_list = process_nodes(json_input)
print(extracted_list)
```

```python
def create_node_system(json_data):
    nodes = json_data['drawflow']['Home']['data']
    node_system = {}

    for node_id, node_info in nodes.items():
        # Initialize the node system dictionary for each node
        node_system[node_id] = {
            "type": node_info['class'],
            "headNodeId": [],
            "childNodeId": [],
            "internalData": node_info['data']
        }

    # Populate headNodeId and childNodeId based on connections
    for node_id, node_info in nodes.items():
        # Handle outputs to determine child nodes
        if 'outputs' in node_info:
            for output_key, output_value in node_info['outputs'].items():
                for connection in output_value['connections']:
                    child_node_id = connection['node']
                    node_system[node_id]['childNodeId'].append(child_node_id)
                    # Also update the head node information for the connected node
                    if node_id not in node_system[child_node_id]['headNodeId']:
                        node_system[child_node_id]['headNodeId'].append(node_id)

    return node_system

# Example usage:
node_system_output = create_node_system(json_input)
print(node_system_output)
```

```python
def create_node_system(json_data):
    nodes = json_data['drawflow']['Home']['data']
    node_system = {}

    for node_id, node_info in nodes.items():
        # Initialize the node system dictionary for each node
        # Ensure internalData points to the innermost 'data' dictionary if it exists
        internal_data = node_info['data']['data'] if 'data' in node_info['data'] else {}
        node_system[node_id] = {
            "type": node_info['class'],
            "headNodeId": [],
            "childNodeId": [],
            "internalData": internal_data
        }

    # Populate headNodeId and childNodeId based on connections
    for node_id, node_info in nodes.items():
        # Handle outputs to determine child nodes
        if 'outputs' in node_info:
            for output_key, output_value in node_info['outputs'].items():
                for connection in output_value['connections']:
                    child_node_id = connection['node']
                    node_system[node_id]['childNodeId'].append(child_node_id)
                    # Also update the head node information for the connected node
                    if node_id not in node_system[child_node_id]['headNodeId']:
                        node_system[child_node_id]['headNodeId'].append(node_id)

    return node_system

# Example usage:
node_system_output = create_node_system(json_input)
print(node_system_output)
```

```python
import json

def generate_code(node_system):
    code = []  # Initialize code as an empty list
    global_variables = {}

    # Extract global variables
    for node_id, node_data in node_system.items():
        if node_data['type'] == 'globalVariable':
            key = node_data['internalData']['key']
            value = node_data['internalData']['valueHolder']
            method = node_data['internalData']['method']
            global_variables[key] = (value, method)

    # Generate global variable assignments
    for key, (value, method) in global_variables.items():
        if method == 'string':
            code.append(f"{key} = '{value}'\n")
        else:
            code.append(f"{key} = {value}\n")

    # Find the starting node
    start_node_id = None
    for node_id, node_data in node_system.items():
        if node_data['type'] == 'fbWebhook' and not node_data['headNodeId']:
            start_node_id = node_id
            break

    # Generate code for each node
    visited_nodes = set()
    def generate_node_code(node_id, indent="", head_node_id=None):
        if node_id in visited_nodes:
            return
        visited_nodes.add(node_id)

        node_data = node_system[node_id]
        node_type = node_data['type']

        if node_type == 'llmGenerate':
            system_prompt = node_data['internalData']['systemPrompt']
            user_prompt = node_data['internalData']['userPrompt']
            code_line = f"{indent}llmGenerate('{system_prompt}', '{user_prompt}', {head_node_id})\n"
            code_line = f"{indent}node{node_id} = " + code_line
            code.append(code_line)
            for child_node_id in node_data['childNodeId']:
                generate_node_code(child_node_id, indent, f"node{node_id}")

        elif node_type == 'extractOutput':
            head_node_id = node_data['headNodeId'][0]
            code_line = f"{indent}extractOutput(node{head_node_id})\n"
            code_line = f"{indent}node{node_id} = " + code_line
            code.append(code_line)
            for child_node_id in node_data['childNodeId']:
                generate_node_code(child_node_id, indent, f"node{node_id}")

        elif node_type == 'switch':
            code.append(f"{indent}match node{node_data['headNodeId'][0]}:\n")
            for child_node_id in node_data['childNodeId']:
                generate_node_code(child_node_id, indent + "    ", f"node{node_data['headNodeId'][0]}")

        elif node_type == 'switchCase':
            input_value = node_data['internalData']['input']
            code.append(f"{indent}case '{input_value}':\n")
            for child_node_id in node_data['childNodeId']:
                generate_node_code(child_node_id, indent + "    ", head_node_id)

    generate_node_code(start_node_id)
    return "\n".join(code)

# Example usage
generated_code = generate_code(node_system_output)
print(generated_code)

# Write to file
# with open('generated_code.py', 'w') as f:
#     f.write(generated_code)
```
